# MANTENIMIENTO MALLA VIAL EN BOGOTÁ

### INTRODUCCION: ANÁLISIS DE TEXTO

```python
import nltk
nltk.download()
```

```python
#Descargar y abrir archivos del Plan Distrital de Planeación en la sección de Movilidad en Bogotá
with open('/Users/lauratovar/Documents/PROYECTO MCPP/movilidad_bogota.txt') as f:
    PDD=f.read() 
```

```python
print(PDD)
```

```python
all_tokens = PDD.split()
all_tokens
```

```python
# Dejar caraceres alfabéticos
alpha = [w for w in all_tokens if w.isalpha()]
alpha
```

```python
len(alpha) #total numero de tokens alfabeticos
```

```python
# Todo en minúsculas
lower = [w.lower() for w in alpha]
lower
```

```python
freqdist = nltk.FreqDist(lower)
freqdist.most_common()
```

```python
from nltk.corpus import stopwords
#Eliminamos preposiciones
stopwords_esp = stopwords.words('spanish')
stopwords_esp
```

```python
# Filter list of all tokens based on list of words of interest
tokens_no_stop = [w for w in lower if w not in stopwords_esp]
tokens_no_stop
```

```python
freqdist = nltk.FreqDist(tokens_no_stop)
freqdist.most_common()
```

```python
tokens_of_interest = [w for w in lower if w in tokens_no_stop]
tokens_of_interest
```

```python
#texto
tokens_PDD = nltk.Text(tokens_of_interest)
tokens_PDD
```

```python
#palabras mas comunes
fdist1= nltk.FreqDist(tokens_PDD)
fdist1.most_common(50)
```

```python
plt.rcParams["figure.figsize"] = [15, 6]
tokens_PDD.dispersion_plot(['obras','vias','vial','localidades','barrios','infraestructura','transporte','calidad','bicicleta','ciclistas','metro','mujeres','tiempo','movilidad','desplazamiento'])
```

```python
tokens_PDD.concordance('vias')
```

```python
tokens_PDD.concordance('calidad')
```

```python
tokens_PDD.concordance('desplazamiento')
```

```python
tokens_PDD.concordance('movilidad')
```
